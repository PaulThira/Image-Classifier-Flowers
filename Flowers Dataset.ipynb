```python
images = [Image.open(os.path.join(image_dir, img_file)) for img_file in image_files]
image_arrays = [np.array(image) for image in images]

# Calculate pixel intensity statistics for each image
image_stats = [np.mean(image_array) for image_array in image_arrays]

# Convert to a numpy array for easier calculation
image_stats = np.array(image_stats)

# Calculate quartiles and IQR
Q1 = np.percentile(image_stats, 5)
Q3 = np.percentile(image_stats, 95)
IQR = Q3 - Q1

# Determine outlier thresholds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Detect outlier images
outlier_indices = np.where((image_stats < lower_bound) | (image_stats > upper_bound))[0]
outlier_images = [images[idx] for idx in outlier_indices]

# Display outlier images
plt.figure(figsize=(12, 8))
num_outliers = len(outlier_images)
for i, outlier_image in enumerate(outlier_images):
    plt.subplot(1, num_outliers, i + 1)
    plt.imshow(outlier_image)
    plt.title(f"Outlier {i+1}")
    plt.axis('off')
plt.tight_layout()
plt.show()

# Print information about detected outliers
print("Detected Outlier Images:")
for i, idx in enumerate(outlier_indices):
    print(f"Outlier {i+1}: Image file name: {image_files[idx]}, Mean Pixel Intensity: {image_stats[idx]:.2f}")
```
